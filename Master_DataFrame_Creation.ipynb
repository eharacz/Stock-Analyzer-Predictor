# Time Series Data from Yahoo!

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2019, 12, 31)
```

```python
start
```

```python
def data_pull(tick):
    df = web.DataReader(f'{tick}', 'yahoo', start, end)
    df['Company']= f'{tick}'
    return df
```

```python
stock_df_list = []
for t in ticker_list: 
    print(t)
    tick = f'{t}'
    stock = data_pull(f'{t}')
    stock_df_list.append(stock)
    time.sleep(1) 
```

# Stock Time Series DataFrame Creation

```python
def master_time_series(df_list):
    """Creates a master stock time series with uncleaned stock dataframes"""
    clean_stock_df_list = []
    for f in df_list:
        f = f.drop(columns=['High', 'Low', 'Open', 'Close', 'Volume'])
        f = f.rename(columns={'Adj Close': f.Company[0]})
        f = f.drop(columns='Company')
        clean_stock_df_list.append(f)
    master_df = pd.concat(clean_stock_df_list, axis=1)
    return master_df
```

```python
master_df = master_time_series(stock_df_list)
```
